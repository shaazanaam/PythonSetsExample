```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
##plt.plot(x,y)
plt.plot(x,y,color='red',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=10)    
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My first graph')
plt.show()

```
